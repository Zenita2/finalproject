```python
#this function that takes an unprocessed tweet from the data list
#and returns all the info plus english and spanish sentiment in a plain text file
def english_and_spanish_sentiment(rawtweet):
    engmodelcorrect = 0
    espmodelcorrect = 0
    bothmodelscorrect = 0
    onlyengcorrect = 0
    onlyespcorrect = 0
    lines = rawtweet.split('\n')
    meta = lines[0].split('\t')
    tweet = [(meta[1], meta[2])]
    for line in lines[1:]:
        word, tag = line.split('\t')
        tweet.append((word,tag))
    tweetdf = pandas.DataFrame(tweet)
    tweetdf = tweetdf.transpose()
    tweetdf.index = ['Word', 'Tag']
    tweettext = tweetdf.loc['Word'].str.cat(sep = ' ')
#this will only work if you've already run the pysentimiento cells above
    espsentiment = espanalyzer.predict(tweettext)
    engsentiment = enganalyzer.predict(tweettext)
    engsentiment = str(engsentiment)
    espsentiment = str(espsentiment)
    if (meta[2] == 'positive' and 'output=POS' in engsentiment
       ) or (meta[2] == 'neutral' and 'output=NEU' in engsentiment
        ) or (meta[2] == 'negative' and 'output=NEG' in engsentiment):
        engmodelcorrect = 1
    if (meta[2] == 'positive' and 'output=POS' in espsentiment
       ) or (meta[2] == 'neutral' and 'output=NEU' in espsentiment
        ) or (meta[2] == 'negative' and 'output=NEG' in espsentiment):
        espmodelcorrect = 1
    if engmodelcorrect == 1 and espmodelcorrect == 1:
        bothmodelscorrect = 1
    elif engmodelcorrect == 1 and espmodelcorrect == 0:
        onlyengcorrect = 1
    elif engmodelcorrect == 0 and espmodelcorrect == 1:
        onlyespcorrect = 1
    with open('englishspanishresults.txt', 'a', encoding = 'utf-8') as f:
        dfstring = tweetdf.to_string(header=False, index=False)
        f.write(dfstring)
        print(f'\nSpanish model: {espsentiment}', file = f)
        print(f'English model: {engsentiment}', file = f)
    return engmodelcorrect, espmodelcorrect, bothmodelscorrect, onlyengcorrect, onlyespcorrect
```

```python
engmodelcount = 0
espmodelcount = 0
bothmodelscount = 0
onlyengcount = 0
onlyespcount = 0
for item in datalist:
    try:
        engmodelcorrect, espmodelcorrect, bothmodelscorrect, onlyengcorrect, onlyespcorrect = english_and_spanish_sentiment(item)
        engmodelcount += engmodelcorrect
        espmodelcount += espmodelcorrect
        bothmodelscount += bothmodelscorrect
        onlyengcount += onlyengcorrect
        onlyespcount += onlyespcorrect
    except Exception as error:
        print(f'Error for index {item}. Details: {error}')
print(f'English model correct: {engmodelcount}')
print(f'Spanish model correct: {espmodelcount}')
print(f'Both models correct: {bothmodelscount}')
print(f'Only English model correct: {onlyengcount}')
print(f'Only Spanish model correct: {onlyespcount}')
```

```python
#this is a modification of the original function
#it doesn't create a text file but it does give us the stats to calculate precision and recall
def english_and_spanish_pr(rawtweet):
    engtruepos = 0
    engfalseposneut = 0
    engfalseposneg = 0
    engfalseneutpos = 0
    engtrueneut = 0
    engfalseneutneg = 0
    engfalsenegpos = 0
    engfalsenegneut = 0
    engtrueneg = 0
    esptruepos = 0
    espfalseposneut = 0
    espfalseposneg = 0
    espfalseneutpos = 0
    esptrueneut = 0
    espfalseneutneg = 0
    espfalsenegpos = 0
    espfalsenegneut = 0
    esptrueneg = 0
    lines = rawtweet.split('\n')
    meta = lines[0].split('\t')
    tweet = [(meta[1], meta[2])]
    for line in lines[1:]:
        word, tag = line.split('\t')
        tweet.append((word,tag))
    tweetdf = pandas.DataFrame(tweet)
    tweetdf = tweetdf.transpose()
    tweetdf.index = ['Word', 'Tag']
    tweettext = tweetdf.loc['Word'].str.cat(sep = ' ')
#this will only work if you've already run the pysentimiento cells above
    espsentiment = espanalyzer.predict(tweettext)
    engsentiment = enganalyzer.predict(tweettext)
    engsentiment = str(engsentiment)
    espsentiment = str(espsentiment)
#english precision and recall numbers
    if 'output=POS' in engsentiment:
        if meta[2] == 'positive':
            engtruepos = 1
        elif meta[2] == 'neutral':
            engfalseposneut = 1
        elif meta[2] == 'negative':
            engfalseposneg = 1
    elif 'output=NEU' in engsentiment:
        if meta[2] == 'positive':
            engfalseneutpos = 1
        elif meta[2] == 'neutral':
            engtrueneut = 1
        elif meta[2] == 'negative':
            engfalseneutneg = 1
    elif 'output=NEG' in engsentiment:
        if meta[2] == 'positive':
            engfalsenegpos = 1
        elif meta[2] == 'neutral':
            engfalsenegneut = 1
        elif meta[2] == 'negative':
            engtrueneg = 1
#spanish precision and recall numbers
    if 'output=POS' in espsentiment:
        if meta[2] == 'positive':
            esptruepos = 1
        elif meta[2] == 'neutral':
            espfalseposneut = 1
        elif meta[2] == 'negative':
            espfalseposneg = 1
    elif 'output=NEU' in espsentiment:
        if meta[2] == 'positive':
            espfalseneutpos = 1
        elif meta[2] == 'neutral':
            esptrueneut = 1
        elif meta[2] == 'negative':
            espfalseneutneg = 1
    elif 'output=NEG' in espsentiment:
        if meta[2] == 'positive':
            espfalsenegpos = 1
        elif meta[2] == 'neutral':
            espfalsenegneut = 1
        elif meta[2] == 'negative':
            esptrueneg = 1
#yes this hurts to look at. but it does work T_T
    return engtruepos, engfalseposneut, engfalseposneg, \
        engfalseneutpos, engtrueneut, engfalseneutneg, \
        engfalsenegpos, engfalsenegneut, engtrueneg, \
        esptruepos, espfalseposneut, espfalseposneg, \
        espfalseneutpos, esptrueneut, espfalseneutneg, \
        espfalsenegpos, espfalsenegneut, esptrueneg
```

```python
#the data is already tagged for language identification
#so one way to run a mixed model would be to only run each language model on the words from that language
def mixed_model_1(rawtweet):
    lines = rawtweet.split('\n')
    meta = lines[0].split('\t')
    tweet = [(meta[1], meta[2])]
    for line in lines[1:]:
        word, tag = line.split('\t')
        tweet.append((word,tag))
    tweetdf = pandas.DataFrame(tweet)
    tweetdf = tweetdf.transpose()
    tweetdf.index = ['Word', 'Tag']
    tweettext = tweetdf.loc['Word'].str.cat(sep = ' ')
    tweettags = tweetdf.loc['Tag']
    tagcounts = tweettags.value_counts()
    engtagcount = tagcounts.get('lang1', 0)
    esptagcount = tagcounts.get('lang2', 0)
#note: this version defaults to english if there's a perfect balance of english and spanish
    if esptagcount > engtagcount:
        sentiment = espanalyzer.predict(tweettext)
        sentmodelused = 'Spanish'
    else:
        sentiment = enganalyzer.predict(tweettext)
        sentmodelused = 'English'
    with open('mixedmodel1results.txt', 'a', encoding = 'utf-8') as f:
        dfstring = tweetdf.to_string(header=False, index=False)
        f.write(dfstring)
        print(f'\n{sentiment}\nModel used: {sentmodelused}', file = f)
    mix1truepos = 0
    mix1falseposneut = 0
    mix1falseposneg = 0
    mix1falseneutpos = 0
    mix1trueneut = 0
    mix1falseneutneg = 0
    mix1falsenegpos = 0
    mix1falsenegneut = 0
    mix1trueneg = 0
    sentiment = str(sentiment)
    if 'output=POS' in sentiment:
        if meta[2] == 'positive':
            mix1truepos = 1
        elif meta[2] == 'neutral':
            mix1falseposneut = 1
        elif meta[2] == 'negative':
            mix1falseposneg = 1
    elif 'output=NEU' in sentiment:
        if meta[2] == 'positive':
            mix1falseneutpos = 1
        elif meta[2] == 'neutral':
            mix1trueneut = 1
        elif meta[2] == 'negative':
            mix1falseneutneg = 1
    elif 'output=NEG' in sentiment:
        if meta[2] == 'positive':
            mix1falsenegpos = 1
        elif meta[2] == 'neutral':
            mix1falsenegneut = 1
        elif meta[2] == 'negative':
            mix1trueneg = 1
    return mix1truepos, mix1falseposneut, mix1falseposneg, \
        mix1falseneutpos, mix1trueneut, mix1falseneutneg, \
        mix1falsenegpos, mix1falsenegneut, mix1trueneg
```

```python
#mixed model 2 is exactly the same as mixed model 1 except it defaults to spanish if there's a language balanca
#since most of the tweets aren't a perfect balance, this shouldn't make much of a difference for the results
#but for the sake of thoroughness, let's test it out
def mixed_model_2(rawtweet):
    lines = rawtweet.split('\n')
    meta = lines[0].split('\t')
    tweet = [(meta[1], meta[2])]
    for line in lines[1:]:
        word, tag = line.split('\t')
        tweet.append((word,tag))
    tweetdf = pandas.DataFrame(tweet)
    tweetdf = tweetdf.transpose()
    tweetdf.index = ['Word', 'Tag']
    tweettext = tweetdf.loc['Word'].str.cat(sep = ' ')
    tweettags = tweetdf.loc['Tag']
    tagcounts = tweettags.value_counts()
    engtagcount = tagcounts.get('lang1', 0)
    esptagcount = tagcounts.get('lang2', 0)
    if engtagcount > esptagcount:
        sentiment = enganalyzer.predict(tweettext)
        sentmodelused = 'English'
    else:
        sentiment = espanalyzer.predict(tweettext)
        sentmodelused = 'Spanish'
    with open('mixedmodel2results.txt', 'a', encoding = 'utf-8') as f:
        dfstring = tweetdf.to_string(header=False, index=False)
        f.write(dfstring)
        print(f'\n{sentiment}\nModel used: {sentmodelused}', file = f)
    mix2truepos = 0
    mix2falseposneut = 0
    mix2falseposneg = 0
    mix2falseneutpos = 0
    mix2trueneut = 0
    mix2falseneutneg = 0
    mix2falsenegpos = 0
    mix2falsenegneut = 0
    mix2trueneg = 0
    sentiment = str(sentiment)
    if 'output=POS' in sentiment:
        if meta[2] == 'positive':
            mix2truepos = 1
        elif meta[2] == 'neutral':
            mix2falseposneut = 1
        elif meta[2] == 'negative':
            mix2falseposneg = 1
    elif 'output=NEU' in sentiment:
        if meta[2] == 'positive':
            mix2falseneutpos = 1
        elif meta[2] == 'neutral':
            mix2trueneut = 1
        elif meta[2] == 'negative':
            mix2falseneutneg = 1
    elif 'output=NEG' in sentiment:
        if meta[2] == 'positive':
            mix2falsenegpos = 1
        elif meta[2] == 'neutral':
            mix2falsenegneut = 1
        elif meta[2] == 'negative':
            mix2trueneg = 1
    return mix2truepos, mix2falseposneut, mix2falseposneg, \
        mix2falseneutpos, mix2trueneut, mix2falseneutneg, \
        mix2falsenegpos, mix2falsenegneut, mix2trueneg
```

```python
mix2precisionpos = mix2trueposcount/(mix2trueposcount + mix2falseposneutcount + mix2falseposnegcount)
mix2precisionneut = mix2trueneutcount/(mix1falseneutposcount + mix2trueneutcount + mix2falseneutnegcount)
mix2precisionneg = mix2truenegcount/(mix2falsenegposcount + mix2falsenegneutcount + mix2truenegcount)
mix2macroprecision = (mix2precisionpos + mix2precisionneut + mix2precisionneg)/3
print(f'Mixed model 2 positive precision: {mix2precisionpos}')
print(f'Mixed model 2 neutral precision: {mix2precisionneut}')
print(f'Mixed model 2 negative precision: {mix2precisionneg}')
print(f'Macroaverage mixed model 2 precision: {mix2macroprecision}')
```

```python
mix2recallpos = mix2trueposcount/(mix2trueposcount + mix2falseneutposcount + mix2falsenegposcount)
mix2recallneut = mix2trueneutcount/(mix2falseposneutcount + mix2trueneutcount + mix2falsenegneutcount)
mix2recallneg = mix2truenegcount/(mix2falseposnegcount + mix2falseneutnegcount + mix2truenegcount)
mix2macrorecall = (mix2recallpos + mix2recallneut + mix2recallneg)/3
print(f'Mixed model 2 positive recall: {mix2recallpos}')
print(f'Mixed model 2 neutral recall: {mix2recallneut}')
print(f'Mixed model 2 negative recall: {mix2recallneg}')
print(f'Macroaverage mixed model 2 recall: {mix2macrorecall}')
```

```python
mix2balancedf1 = 2 * (mix2macroprecision * mix2macrorecall)/(mix2macroprecision + mix2macrorecall)
print(f'Mixed model 2 balanced F1 score: {mix2balancedf1}')
```
